```python
#Let's plotting a few functions at once
y1 = np.sin(x)
y2 = np.cos(x)

x_ticks = [0, np.pi/2, np.pi, 3*np.pi/2, 2*np.pi]
#for labels with TeX notation, use an r (for raw string literal) before the string
x_labels = ['0', r'$\frac{\pi}{2}$', r'$\pi$', r'$\frac{3\pi}{2}$', r'$2\pi$']

#This will add labels, axis limits and a legend:
plt.figure(4)
plt.plot(x, y1, label = 'sin(x)')
plt.plot(x, y2, label = 'cos(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Functions of x')
#plt.legend()  
#Python will attempt to place your legend where it won't interfere with what you're plotting
#You can also move the legend around manually with:
plt.legend(bbox_to_anchor=(1,0.9), prop={'size': 12}) #specify legend position and font size
plt.xlim(0,2*np.pi)
plt.ylim(-1.1, 1.1)
#We can also define our own axes - comment out the x axis array above
plt.xticks(x_ticks, x_labels)
plt.show()

#For presentations, it will be helpful to adjust your label sizes
#Scroll to the box above and try adjusting those parameters
```

```python
#Some more things that matplotlib.pyplot can do:

#First, we're going to generate some mock data from a Gaussian distribution:
mu, sigma = 0, 5
#s = np.random.normal(mu, sigma, 100)
#s = np.random.normal(mu, sigma, 1000)
s = np.random.normal(mu, sigma, 100000)
#print(s)

#matplotlib has a histogram function you can use, if you specify:
# 1) the array you're binning
# 2) the number of bins or your bin edges 

#plt.figure(4)
#counts, bins, ignore = plt.hist(x, 30, normed= 'True') #bin x in 30 evenly spaced bins, and normalize the histogram
#plt.show()

#plt.hist returns the following:
#print(counts) #An array that lists the value in each bin
#print(bins) #The generated bin edges

#Let's overplot a gaussian function on this, to confirm this is the best fit for our data
#Define the gaussian probability density function
def p(x, mu, sigma):
    return (1/(np.sqrt(2*np.pi*sigma**2)))*np.exp((-(x - mu)**2)/(2*sigma**2))

plt.figure(8)
counts, bins, ignore = plt.hist(s, 40, normed= 'True') #bin x in 30 evenly spaced bins, and normalize the histogram
plt.plot(bins, p(bins, mu, sigma), 'r-')   
plt.show()

#This doesn't look great with only 100 points, so scroll back up and try changing the number of data points we're generating...


 

```

```python
#Another thing that could be useful is subplots:

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure(1)
ax = plt.subplot(221) #2x2 figures, 1st figure
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')
#Since some of these plots share axes, we can turn them off:
plt.setp(ax.get_xticklabels(), visible=False)

ax = plt.subplot(222) 
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.setp(ax.get_xticklabels(), visible=False)
plt.setp(ax.get_yticklabels(), visible=False)

ax = plt.subplot(223) #2x2 figures, 1st figure
plt.plot(t1, 2*f(t1), 'bo', t2, 2*f(t2), 'k')

ax = plt.subplot(224) 
plt.plot(t2, 2*np.cos(2*np.pi*t2), 'r--')
plt.setp(ax.get_yticklabels(), visible=False)

plt.show()
```
