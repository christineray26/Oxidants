```python
#This is just importing various modules and adjusting settings for figures

import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
from scipy.integrate import odeint
plt.rcParams['axes.titlesize'] = 22
plt.rcParams['axes.labelsize'] = 20  
plt.rcParams['xtick.labelsize'] = 18
plt.rcParams['ytick.labelsize'] = 18
plt.rcParams['axes.titlepad'] = 40 
plt.rcParams['axes.labelpad'] = 16 
```

```python
#Constants, for Ca2+ at 25 degrees C
A=0.5085
B=0.3281
a=6
z=2

#Values for ionic strength:
I = np.logspace(-4, 0, 40)
#40 values between -4 and 0, 
#which corresponds to increments of 0.1 in logspace

#-log gamma values: 
DHcond_log = A*(z**2)*np.sqrt(I) #condensed
DHfull_log = DHcond_log/(1 + a*B*np.sqrt(I)) #full
Davies_log = A*(z**2)*((np.sqrt(I)/(1+np.sqrt(I)))-0.2*I) #Davies

#gamma = 10^(-log gamma):
DHcond = 10**-DHcond_log #condensed
DHfull = 10**-DHfull_log #full
Davies = 10**-Davies_log #Davies
```

```python
#Plotting
fig1 = plt.figure(1)
fig1.set_figheight(6.5)
fig1.set_figwidth(11)
plt.loglog(I, DHcond, label = 'D-H condensed')
plt.loglog(I, DHfull, label = 'D-H full')
plt.loglog(I, Davies, label = 'Davies')
plt.xlim(10**-4, 10)
plt.ylim(10**-2.3,1)
plt.ylabel('Activity Coefficient ($\gamma$)')
plt.xlabel('Ionic Strength')
plt.title('Activity Coefficient of Ca$^{2+}$')
plt.legend(bbox_to_anchor=(0.35,0.3), prop={'size': 16})
plt.grid(True, which="both")
plt.show()
```

```python
#Constants for different ions:
a_Na = 4.25 #choosing the average value
a_Cl = 3
a_SO4 = 4.25
z_Na = 1
z_Cl = 1
z_SO4 = 2

#-log activity
Na_log = A*(z_Na**2)*np.sqrt(I)/(1 + a_Na*B*np.sqrt(I)) #Na2+
Cl_log = A*(z_Cl**2)*np.sqrt(I)/(1 + a_Cl*B*np.sqrt(I)) #Cl-
SO4_log = A*(z_SO4**2)*np.sqrt(I)/(1 + a_SO4*B*np.sqrt(I)) #SO4 2-

#activity
Na = 10**-Na_log
Cl = 10**-Cl_log
SO4 = 10**-SO4_log
Ca = DHfull #from earlier
```

```python
#Plotting
fig2 = plt.figure(2)
fig2.set_figheight(6.5)
fig2.set_figwidth(11)
plt.loglog(I, Ca, label = 'Ca$^{2+}$')
plt.loglog(I, Na, label = 'Na$^{+}$')
plt.loglog(I, Cl, label = 'Cl$^{-}$')
plt.loglog(I, SO4, label = 'SO$_4^{2-}$')
plt.xlim(10**-4, 10)
plt.ylim(1e-1,1)
plt.ylabel('Activity Coefficient ($\gamma$)')
plt.xlabel('Ionic Strength')
plt.title('Activity Coefficient of Different Ions')
plt.legend(bbox_to_anchor=(0.2,0.4), prop={'size': 16})
plt.grid(True, which="both")
plt.show()
```
