```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integrate
from scipy.integrate import odeint
import scipy.interpolate as interp
plt.rcParams['axes.titlesize'] = 20  
plt.rcParams['axes.labelsize'] = 20  
plt.rcParams['xtick.labelsize'] = 18
plt.rcParams['ytick.labelsize'] = 18
plt.rcParams['axes.titlepad'] = 30 
plt.rcParams['axes.labelpad'] = 16 
```

```python
#x = log[H4SiO4]:
x = np.asarray(np.arange(-5,-1,0.001)) 

#y = log [Na+]/[H+] - determined by the following line equations:

#1. albite-kaolinite
K1 = 0.125
y1 = K1 - 2*x

#2. albite - gibbste
K2 = -4.02
y2 = K2 - 3*x

#3. albite - Na-smectite
K3 = 1.42
y3 = K3 - 1.67*x

#4. Na-smectite - kaolinite
K4 = -7.645
y4 = K4 - 4*x


#The next four reactions are vertical lines, so we will define a generic y:
y = np.asarray(np.arange(4,9,0.001)) #10 numbers evenly incremented in logspace between 10^4 and 10^9


#And define x using the following:

#5. gibbsite - kaolinite
K5 = 8.29
x5 = np.zeros(len(y)) #create an empty array with same number of elements as y
x5[:] = -K5/2 #make all array elements equal to this value

#6. quartz - silicic acid
K6 = -4.00
x6 = np.zeros(len(y)) 
x6[:] = K6

#7. chalcedony - silicic acid
K7 = -3.73
x7 = np.zeros(len(y)) 
x7[:] = K7

#8. amorphous silica - silicic acid
K8 = -2.71
x8 = np.zeros(len(y)) 
x8[:] = K8
```

```python
#Plot the version without any intersection points removed:
#We can use this to figure out where different regions should be by commenting out different lines
fig1 = plt.figure(1)
fig1.set_figheight(9)
fig1.set_figwidth(10)
plt.clf()
plt.plot(x,y1, label = 'albite-kaolinite')
plt.plot(x,y2, label = 'albite - gibbste')
plt.plot(x,y3, label = 'albite - Na-smectite')
plt.plot(x,y4, label = 'Na-smectite - kaolinite')
plt.plot(x5,y, label = 'gibbsite - kaolinite')
plt.plot(x6,y, '--', label = 'quartz')
plt.plot(x7,y, '--', label = 'chalcedony')
plt.plot(x8,y, '--', label = 'amorphous silica')
plt.legend()
plt.xlabel('log[H$_4$SiO$_4$]')
plt.ylabel(r'log$\frac{[{Na}^+]}{[H^+]}$')
plt.xlim(-5,-1)
plt.ylim(4,9)
#plt.plot(x2,y1)
plt.show()
```

```python
#Define a function to calculate intersection point of two lines
def line(p1, p2):
    A = (p1[1] - p2[1])
    B = (p2[0] - p1[0])
    C = (p1[0]*p2[1] - p2[0]*p1[1])
    return A, B, -C

def intersection(x1,y1,x2,y2):
    p1 = [x1[0],y1[0]]
    p2 = [x1[1],y1[1]]
    p3 = [x2[0],y2[0]]
    p4 = [x2[1],y2[1]]
    L1 = line(p1,p2)
    L2 = line(p3,p4)
    D  = L1[0] * L2[1] - L1[1] * L2[0]
    Dx = L1[2] * L2[1] - L1[1] * L2[2]
    Dy = L1[0] * L2[2] - L1[2] * L2[0]
    if D != 0:
        x = np.round((Dx / D), decimals = 3)
        y = np.round((Dy / D), decimals = 3)
        return x,y
    else:
        return False
```

```python
#A demonstration that this works:
#Intersection of albite-kaolinite and albite-gibbsite
x_int,y_int = intersection(x,y1,x,y2)
print(x_int,y_int)
```

```python
#Now calculate all intersection points:

#albite-kaolinite, albite-gibbsite:
x_12,y_12 = intersection(x,y1,x,y2)

#albite-kaolinite, albite-Na-smectite:
x_13, y_13 = intersection(x,y1,x,y3)

#albite-gibbsite, albite-Na-smectite:
x_23, y_23 = intersection(x,y2,x,y3)

#albite-kaolinite, Na-smectite-kaolinite:
x_14, y_14 = intersection(x,y1,x,y4)

#albite-Na-smectite, Na-smectite-kaolinite:
x_34, y_34 = intersection(x,y3,x,y4)

#albite-kaolinite, gibbsite-kaolinite:
x_15, y_15 = intersection(x,y1,x5,y)

#albite-gibbsite, gibbsite-kaolinite:
x_25, y_25 = intersection(x,y2,x5,y)

#Na-smectite-kaolinite, gibbsite-kaolinite:
x_45, y_45 = intersection(x,y4,x5,y)
```

```python
#Setting up the new plot:
fig2 = plt.figure(2)
fig2.set_figheight(9)
fig2.set_figwidth(10)
plt.clf()

#First we'll do the albite-kaolinite-gibbsite sections:
#albite-kaolinite-gibbsite:
#plt.plot(x[x>=x_12],y1[y1<=y_12], label = 'albite-kaolinite') #plotted arrays amended at intersection points
plt.plot(x[x<=x_12],y2[y2>=y_12], 'k') #albite - gibbste 

#This one is trickier because all x values are the same
#We will tell it to chop off the x array after the numberof elements in y, up to the intersection point, have been reached
plt.plot(x5[0:len(y[y<=y_15])], y[y<=y_15], 'k') #gibbsite - kaolinite

#Adding labels:
plt.text(-2.3, 6.5, 'Albite', fontsize = 20)
plt.text(-4, 5.5, 'Kaolinite', fontsize = 20)
plt.text(-4.85, 6.5, 'Gibbsite', fontsize = 20)

#Now add in Na-smectite:
plt.plot(x[x>=x_34],y3[y3<=y_34+0.001], 'k') #albite - Na-smectite
plt.plot(x[x>=x_14],y4[y4<=y_14], 'k') #Na-smectite - kaolinite

#And modify the albite kaolinite line again:
x_AK = x[x_12<=x]
x_AK = x_AK[x_AK<=x_14]
y_AK = y1[y1<=y_12]
y_AK = y_AK[y_14<=y_AK]

plt.plot(x_AK, y_AK, 'k') #albite-kaolinite

#And add another label:
plt.text(-3.1, 5, 'Na-Smectite', fontsize = 20)

#And these are all of our vertical saturation lines
plt.plot(x6,y, '--', label = 'Quartz Saturation')
plt.plot(x7,y, '--', label = 'Chalcedony Saturation')
plt.plot(x8,y, '--', label = 'Amorphous Silica Saturation')

#And finally, the two points in the questions:
plt.plot(-4, (-2.9+7.5), 'b*', markersize = 18, label = 'Question 1')
plt.plot(-4, (0.5+8), 'r*', markersize = 18, label = 'Question 2')

plt.legend(prop={'size': 16})
plt.xlabel('log[H$_4$SiO$_4$]')
plt.ylabel(r'log$\frac{[{Na}^+]}{[H^+]}$')
plt.xlim(-5,-1)
plt.ylim(4,9)
#plt.plot(x2,y1)
plt.show()
```
